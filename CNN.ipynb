```python

import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import TensorBoard
import time
import numpy as np
import datetime
from keras.utils import plot_model


NAME = "CNN-{}".format(int(time.time()))
tensorboard = TensorBoard(log_dir='.\logs\{}'.format(NAME))  

#file_writer = tf.summary.FileWriter('logs', sess.graph)
#file_writer = tf.summary.create_file_writer('logs')

     

mnist = tf.keras.datasets.mnist #default 28by28 number database
(x_train, y_train), (x_test, y_test)=mnist.load_data()
#normalize from 255 to 1
x_train=tf.keras.utils.normalize(x_train,axis=1)
x_test=tf.keras.utils.normalize(x_test,axis=1)
#hiddenlayers+outputlayer
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(10,activation=tf.nn.softmax))

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
# checkpoint

model.fit(x_train, y_train, epochs=3, callbacks=[tensorboard])
#tensorboard --logdir logs/CNN-xxxxx
plot_model(model, to_file='CNN_model.png')

```

```python
#overfitting test
val_loss, val_acc = model.evaluate (x_test,y_test)
print (val_loss, val_acc)
fp=open("mydata.dat","w")
fp.write("#LL Iteration  Accuracy\n")
```
